# Week 7: Line Plot of IoT Sensor Readins Over Time

1. Access your preprocessed data from Homework No. 3.

```python
import pandas as pd


# Load cleaned IoT data and remove duplicate device_id rows, keeping the first occurrence
df = pd.read_csv("data/cleaned_iot_data.csv")
df = df.drop_duplicates(subset="device_id", keep="first")

# Remove degree celsius symbol from data_value column
# (Skipped: data_value is already float, so no string replacement needed)

# Replace all 'temperature' with 'Temparature' in data_type column
df["data_type"] = df["data_type"].replace("temperature", "Temparature")

# Filter to first 20 entries only
df = df.head(20)

# Display first few rows to verify data
print(df.head())
```

2. To ensure proper plotting, convert the timestamp column into a datetime format.

```python
# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

3. Import the required libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

4. Set the visualization style.

```python
sns.set(style="whitegrid")
```

5. Create the line plot.

```python
plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x=df["timestamp"], y=df["data_value"], marker="o")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title("IoT Sensor Readings Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Temperature", fontsize=12)


# Display the plot
plt.show()
```
